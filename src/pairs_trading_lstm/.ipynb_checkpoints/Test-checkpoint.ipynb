```python
import yfinance as yf

tickers = {"Tech": ["AAPL", "MSFT", "GOOG", "NVDA", "ADBE", "ORCL", "INTC", "TXN", "AMD", "PYPL"], 
           "Financials": ["JPM", "BAC", "C", "GS", "MS", "BRK-A", "V", "MA", "AXP", "BLK"], 
           "ConsumerDisc": ["TSLA", "NKE", "HD", "LOW", "SBUX", "MCD", "TGT", "TJX", "LULU", "SONO"], 
           "Industrials": ["GE", "CAT", "BA", "MMM", "HON", "ITW", "UNP", "LMT", "FDX", "RTX"], 
           "Utilities": ["NEE", "DUK", "SO", "D", "AEP", "EXC", "XEL", "SRE", "PEG", "CNP"],
           "Healthcare": ["LLY", "JNJ", "ABBV", "MRK", "PFE", "AMGN", "GILD", "ISRG", "HCA", "UNH"],
           "Energy": ["XOM","CVX","COP","EOG","SLB","OXY","DVN","FANG","LNG","HAL"],
           "ConsumerStaples": ["PG","KO","PEP","WMT","COST","CL","KMB","GIS","HSY","MKC"],
           "RealEstate": ["AMT","PLD","EQIX","O","SPG","WELL","DLR","AVB","EQR","VICI"],
           "Communications": ["META","NFLX","DIS","TMUS","VZ","T","CHTR","PSKY","CMCSA","WBD"]}

data = {}
for sect in tickers.keys():
    rawData = yf.download(tickers[sect], start="2019-01-01", group_by="ticker", auto_adjust="true")
    data[sect] = rawData.xs("Close", axis=1, level=1)
    print(data[sect].head())
```
